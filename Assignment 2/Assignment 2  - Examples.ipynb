```python
import os
import re
import imageio
import assignment_2 as dip
import matplotlib.pyplot as plt
```

```python
img_folder = 'input_and_output_imgs/'
test_cases_folder = [i for i in os.listdir() if re.search('Casos de Teste', i)][0]
```

```python
def print_input_output_image(p, test_case):
    if(len(params) == 5):
        r, f = dip.apply_image_transformation(imageio.imread(img_folder+p[0]), 
                                              int(p[1]), 0, None, float(p[3]), float(p[4]))
    else:
        r, f = dip.apply_image_transformation(imageio.imread(img_folder+p[0]), 
                                              int(p[1]), 0, int(p[3]), float(p[4]), float(p[5]))
        
    plt.figure(figsize=(12,12))
    plt.suptitle('Test Case #'+str(test_case), y=0.85)
    
    plt.subplot(121)
    plt.title('Image before transformation')
    plt.imshow(r, cmap="gray", vmin=0, vmax=255)
    
    plt.subplot(122)
    plt.title('Image after transformation')
    plt.imshow(f, cmap="gray", vmin=0, vmax=255)
    
    plt.tight_layout()
    plt.show()
```

```python
for i in range(len([i for i in os.listdir(test_cases_folder) if re.search('.in', i)])):
    with open(test_cases_folder+'/'+str(i+1)+'.in', 'r') as f:
        params = f.read().split('\n')
        
    print_input_output_image(params, i+1)
```
