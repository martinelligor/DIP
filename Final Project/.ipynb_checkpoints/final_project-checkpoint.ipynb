```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
from scripts.canny import Canny
```

```python
# used to transform image to gray-scale
def rgb2gray(img_rgb):
    return np.floor(np.dot(img_rgb[...,:3], [0.299, 0.587, 0.114]))

# use for a test in image with white background and natural background.
def image_segmentation(image, threshold):
    #todos os pixels recebem 0 inicialmente.
    mask = np.zeros(image.shape)
    #se pixel >= threshold em image, atribui ele como 255.
    mask[image>=threshold] = 255
    #retorna a máscara gerada pela segmentação.
    return mask
```

```python
img_white_bg, img_natural_bg = imageio.imread('images/7.jpg'), imageio.imread('images/7.png')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16,6))

ax[0].imshow(img_white_bg)
ax[1].imshow(img_natural_bg)
```

```python
img_white_bg = rgb2gray(img_white_bg)
img_natural_bg = rgb2gray(img_natural_bg)

fig, ax = plt.subplots(1, 2, figsize=(16,6))

ax[0].imshow(img_white_bg)
ax[1].imshow(img_natural_bg)
```

```python
img_w_aux = image_segmentation(img_white_bg, 240)
img_n_aux = image_segmentation(img_natural_bg, 150)

fig, ax = plt.subplots(1, 2, figsize=(16,6))

ax[0].imshow(img_w_aux)
ax[1].imshow(img_n_aux)
```

```python
canny_white = Canny(img_white_bg, sigma=2, kernel_size=3, 
                    low_threshold=.1, high_threshold=.55, weak_pixel=10).canny()

canny_natural = Canny(img_natural_bg, sigma=2, kernel_size=3, 
                      low_threshold=.1, high_threshold=.45, weak_pixel=30).canny()

fig, ax = plt.subplots(1, 2, figsize=(16,6))

ax[0].imshow(canny_white)
ax[1].imshow(canny_natural)
```
